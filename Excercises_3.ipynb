# Загрузка базы вопросов ЧГК

Нужно загрузить все доступные вопросы и ответы к ним с базы http:///db.chgk.info . Результаты необходимо сохранить в каком-либо формате в форме вопрос-ответ (pickle, sqlite, текстовый файл и т.д.). 

```python
import lxml.html
import lxml.etree
import requests
import re
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
database = pd.DataFrame([],columns = ["Question","Answer"])
```

```python
host_url = 'https://db.chgk.info'


def read_txt_chgk(url):
    file = requests.get(url , verify=False)
    pattern = re.compile(r"Вопрос.*?:\n(.*?)\n\nОтвет:\n(.*?)\n\n",re.DOTALL)
    ans = pattern.findall(file.text)
    temp = pd.DataFrame(ans,columns = ["Question","Answer"])
    return temp

def read_html_chgk(inpoint):
    global database
    global host_url
    #print(inpoint)
    page = requests.get(inpoint,verify=False)
    page_html = lxml.html.fromstring(page.text)
    if page.status_code == 200:
        if page_html.xpath('//div[@id="main"]/p/a/text()').count("[TXT]") == 0:
            links = page_html.xpath('//div[@id="main"]/ul/li/a/@href')
            for el in links:
                read_html_chgk(host_url + el)
        else:
            src_url = host_url + page_html.xpath('//div[@id="main"]/p/a/@href')[1]
            temp_db = read_txt_chgk(src_url)
            #print(src_url,inpoint,page_html.xpath('//div[@id="main"]/p/a/@href'),len(temp_db))
            #print(database)
            database = pd.concat([database,temp_db],ignore_index = True)
    
```

```python
read_html_chgk('https://db.chgk.info/tour')
```
